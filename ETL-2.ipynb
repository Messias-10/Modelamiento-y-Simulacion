```python
df_NL = pd.read_csv('BRORGA2441NL_NAC_2023_2.csv', sep = '|')

df_NL.columns = ['Código Comuna', 'Número de Manzana', 'Número de Predial', 'Número correlativo de la línea de construcción', 'Código del material estructural de la línea de construcción', 'Código de calidad de la línea de construcción', 'Año de la línea de construcción', 'Superficie de la línea de construcción (sin decimales, en m2 o m3 según tipo de construcción)', 'Código de destino de la línea de construcción', 'Código de condición especial de la línea de construcción']

df_NL
```

Dejar las comunas de la gran santiago y colocar el número de rol y manzana con predial

```python
df_NL = df_NL[df_NL['Código Comuna'].isin(comunas.keys())]

df_NL.insert(0, 'Cod', df_NL.apply(lambda row: str(row['Código Comuna']) + '-' + str(row['Número de Manzana']), axis=1))
df_NL.insert(1, 'Rol', df_NL.apply(lambda row: str(row['Número de Manzana']) + '-' + str(row['Número de Predial']), axis=1))

df_NL
```

```python
df_NL.insert(2, 'Comuna', df_NL['Código Comuna'].map(comunas))
df_NL.insert(3, 'Destino de la línea de Construcción', df_NL['Código de destino de la línea de construcción'].map(destinos))
df_NL.insert(4, 'Material Estructural', df_NL['Código del material estructural de la línea de construcción'].map(materiales))
df_NL.insert(5, 'Calidad', df_NL['Código de calidad de la línea de construcción'].map(calidad))

df_NL = df_NL.drop(['Código Comuna', 'Código de destino de la línea de construcción', 'Número de Manzana', 'Número de Predial', 'Código del material estructural de la línea de construcción', 'Código de calidad de la línea de construcción', 'Código de condición especial de la línea de construcción', 'Número correlativo de la línea de construcción'], axis=1)

df_NL
```

Combinación de ambas base de datos

```python
df_BD = pd.merge(df_N, df_NL, on=['Cod', 'Rol'])
df_BD
```

```python
df_BD.drop(['Comuna_y', 'Destino de la línea de Construcción'], axis=1, inplace=True)

df_BD.rename(columns={'Comuna_x': 'Comuna'}, inplace=True)

df_BD.fillna('-', inplace=True)

df_BD
```

```python
df_BD = df_BD.groupby("Rol").agg({
    "Cod": "first",
    "Comuna": "first",
    "Código Comuna": "first",
    "Dirección o nombre del predio": "first",
    "Destino Principal": lambda x:', '.join(set(x)),
    "Avalúo fiscal total": "sum",
    "Contribución semestral (con aseo)": "sum",
    "Avalúo exento de la propiedad": "sum",
    "Superficie total del terreno de la propiedad (sin decimales, en m2)": "sum",
    "Material Estructural": lambda x:', '.join(set(x)),
    "Calidad": lambda x:', '.join(set(x)),
    "Año de la línea de construcción": "first",
    "Superficie de la línea de construcción (sin decimales, en m2 o m3 según tipo de construcción)": "sum",
}).reset_index()

df_BD.to_csv('Data.csv', index=False)

df_BD
```
